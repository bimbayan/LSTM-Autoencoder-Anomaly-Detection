```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load data again
data = pd.read_csv(r"C:\Users\bimba\.vscode\LSTM Autoencoder Project\megamassign\NAB-master\data\realKnownCause\ambient_temperature_system_failure.csv")
data['timestamp'] = pd.to_datetime(data['timestamp'])
data = data.sort_values(by='timestamp')

# Reset index
data = data.reset_index(drop=True)

# Check first few rows
print(data.head())
print("\nData shape:", data.shape)
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Scale the temperature values to range [0, 1]
scaler = MinMaxScaler()
data['value'] = scaler.fit_transform(data['value'].values.reshape(-1, 1))

# Check result
print(data.head())
```

```python
# Create sequences of 30 readings (can adjust later)
TIME_STEPS = 30

def create_sequences(values, time_steps=TIME_STEPS):
    output = []
    for i in range(len(values) - time_steps):
        output.append(values[i : (i + time_steps)])
    return np.stack(output)

# Create sequences
sequences = create_sequences(data['value'].values)

print("Shape of sequences:", sequences.shape)
```

```python
#split into training and testing sets

train_size = int(len(sequences) * 0.8)
train_data = sequences[:train_size]
test_data = sequences[train_size:]

print("Train data shape:", train_data.shape)
print("Test data shape:", test_data.shape)
```

```python
import numpy as np

np.save("train_data.npy", train_data)
np.save("test_data.npy", test_data)
```
